```python
import numpy as np
# Save the nutrion in the dataframe
nutrion = np.array(nutrion )
nutrion = pd.DataFrame(nutrion,columns = ['nutrion'])
```

```python

# code to scrap the discription
for webrec in result1:
  ingredien=[]
  recipi=[]
  request = Request(webrec, headers=headers)
 # url = urllib.request.urlopen(web, headers={'User-Agent': 'Mozilla/5.0'})
  content = urlopen(request).read()
  soup = BeautifulSoup(content, 'lxml')
  # code to scrap the Recipie name
  h1tag=soup.find('h1',attrs={'comp type--lion article-heading mntl-text-block'})
  recipie_name.append(h1tag.text)
  
  #  #code to scrap the discription name
  ptag=soup.find('p',attrs={'comp type--dog article-subheading'})
  discription.append(ptag.text)
  #Code to scrap discription
  for a in soup.findAll('li', attrs={'mntl-structured-ingredients__list-item '}):
    ingre=a.find('p')
    ingredien.append(ingre.text)
  ingredient.append(ingredien)
  # #code to scrap rcipie steps
  for a in soup.findAll('li', attrs={'comp mntl-sc-block-group--LI mntl-sc-block mntl-sc-block-startgroup'}):
    step=a.find('p')
  # print(step.text)
    recipi.append(step.text)
  recipie.append(recipi)
```

```python
ingredient = np.array(ingredient )
recipie_name = np.array(recipie_name )
discription = np.array(discription )
recipie = np.array(recipie )
```

```python
import pandas as pd
Recipie=pd.read_csv('/content/FinalRecipieData.csv', on_bad_lines='skip')
```

```python
import re
Recipie['recipe_name']=Recipie['recipe_name'].str.lower()
# # Proprocessing the data
# Code to remove the Hashtags from the text
Recipie['recipe_name']=Recipie['recipe_name'].apply(lambda x:re.sub(r'\B#\S+','',str(x)))
# Code to remove the links from the text
Recipie['recipe_name']=Recipie['recipe_name'].apply(lambda x:re.sub(r"http\S+", "", x))
# Code to remove the Special characters from the text 
Recipie['recipe_name']=Recipie['recipe_name'].apply(lambda x:' '.join(re.findall(r'\w+', x)))
# Code to substitute the multiple spaces with single spaces
# Recipie['recipe_name']=Recipie['recipe_name'].apply(lambda x:re.sub(r'\s+', ' ', x, flags=re.I))
# Code to remove all the single characters in the text
Recipie['recipe_name']=Recipie['recipe_name'].apply(lambda x:re.sub(r'\s+[a-zA-Z]\s+', '', x))
# Remove the twitter handlers
Recipie['recipe_name']=Recipie['recipe_name'].apply(lambda x:re.sub('@[^\s]+','',x))
Recipie['recipe_name']
```
